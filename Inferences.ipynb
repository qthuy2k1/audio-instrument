```python
from transformers import pipeline

classifier = pipeline("audio-classification", model="qthuy2k1/audio_instrument")
```

```python
from datasets import load_dataset, Audio

test_file = load_dataset("audiofolder", data_dir="/content/drive/MyDrive/test_file", split="train")
test_file = test_file.cast_column("audio", Audio(sampling_rate=16000))
```

```python
print(test_file[12])
```

```python
for x in test_file:
  print(x["audio"]["path"])
  print(classifier(x["audio"]["path"]))
  print("\n")
```
